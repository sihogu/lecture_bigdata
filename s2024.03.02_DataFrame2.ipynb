```python
import pandas as pd
```

```python
exam_data = {'수학':[ 90, 80,  70, 60,  70],
             '영어':[ 98, 89,  99, 90, 100],
             '음악':[ 85, 95, 100, 90,  90],
             '체육':[100, 90,  90, 80,  70]}

df = pd.DataFrame(exam_data, index = ['서준', '우현', '인아', '진수', '진희'])
df
```

```python
exam_data2 = {'이름':['서준', '우현', '인아', '진수', '진희'],
              '수학':[ 90, 80,  70, 60,  70],
              '영어':[ 98, 89,  99, 90, 100],
              '음악':[ 85, 95, 100, 90,  90],
              '체육':[100, 90,  90, 80,  70]}

df2 = pd.DataFrame(exam_data2)

# DF.set_index('col명', inplace = False) : 특정 columns(열)을 index로 설정한다.
df2.set_index('이름', inplace = True)
df2
```

```python
# 열(columns) -> 행(row)     : DF['col명']['idx명']  ,  DF.col명['idx명']
# 행(row)     -> 열(columns) : DF.loc['idx명', 'col명'] , DF.loc['idx명']['col명']
#                              DF.iloc[idx번호, col번호]
print(df['영어']['인아'])
print(df.영어['인아'])
print(df.loc['진수']['영어'])
print(df.loc['진수', '영어'])
print(df.iloc[3, 1])
```

```python
# DF.loc['idx명', 'col명']                                      -> 단일 원소
# DF.loc[['idx명1', 'idx명2', ...], ['col명1', 'col명2', ...]]  -> 다수 원소들
df.loc[['서준', '진희', '인아'], ['수학', '영어']]
```

```python
# DF.iloc[idx번호, col번호]                                      : 단일 원소
# DF.iloc[[idx번호1, idx번호2, ...], [col번호1, col번호2, ...]]  : 다수 원소
df.iloc[[0, -1], [0, -1]]
```

```python
# DF.loc[['idx명1','idx명2,'...]]
# DF.loc['idx명1':'idx명2']       : idx명1부터 idx명2까지 조회 [이상:이하]
# DF.loc[idx번호1:idx번호2]       : idx번호1부터 idx명2전까지  [이상:미만]
df.loc[['서준', '우현', '인아']]
```

```python
df4 = pd.read_csv('csv/class_1.csv', encoding = 'euckr',
                  names = ['name', 'addr', 'kor', 'eng', 'math', 'sci', 'his'])
# DF.set_index('col명', inplace = False/True) : 특정한 columns을 index(축)으로 지정
#              'col명'                        - 단일 index 지정
#             ['col명1', 'col명'2]            - 다중 index 지정
# set_index() 함수는 새로운 객체를 반환, 원본이 수정되기를 원하면
# DF = DF.set_index()  ,  DF.set_index(inplace = True)
df4 = df4.set_index(['name'])
df4
```

```python
df5 = df4.copy()
new_index = ['유재석', '노홍철', '박명수', '황광희', '정준하', '하하', '길']
# DF.reindex([새로운 index배열]) : 새로운 index배열에 기존 index와 동일한 원소가
#                                  있다면 그 값들을 채우고, 없다면 NaN(Not a Number)
#                                  으로 빈 칸을 채운다.
#                                  기존의 index들을 재배치하는 목적이 있다.
# NaN(Not a Number) 은 결측치를 나태내며, numpy.nan 에서 유래됐고, 실수형 데이터이다.
df5.reindex(new_index)
```

```python
df5 = df4.copy()
# DF.reset_index()  : index(축)을 초기화한다. index(축)의 데이터들을 columns으로 이동
df5.reset_index()
```

```python
df5 = df4.copy()
# DF.sort_index()   : index명을 기준으로 오름차순/내림차순으로 정렬
#                     ascending  = True(기본) / False
df5.sort_index(ascending = False, inplace = True)
df5
```

```python
df5 = df4.copy()
# DF.sort_values(by = [기준 col명들])  : 특정한 columns의 값을 기존으로
#                                        오름차순/내림차순으로 정렬
#                                        ascending = True(기본) / False
df5.sort_values(by = ['kor', 'eng'], ascending = False)
```

```python
data = {'국어':100, '영어':80, '수학':90}
sr = pd.Series(data)

# Series + 연산자(+,-,*,/) + 숫자 : Series의 모든 원소(값)들을 동일한 1개의 숫자 연산
print(sr)
print()
print(sr + 10)
```

```python
data1 = {'국어':100, '영어':80, '수학':90}
data2 = {'수학':85, '국어':95, '체육': 89}
sr1 = pd.Series(data1)
sr2 = pd.Series(data2)
print(sr1)
print()
print(sr2)
# Series + 연산자(+,-,*,/) + Series : 나열된 순서와 상관없이 동일 index명끼리
#                                     연산이 이루어지고, 서로 없는 index끼리는
#                                     연산이 이루어지지 않고 NaN으로 대체
sr1 + sr2
```

```python
# Series.연산메서드(Series, fill_value = 값) : NaN(결측치)에 대해 대체값으로
#                                              연산도 가능한다.
# add(+), sub(-), mul(*), div(/)
print(sr1 + sr2)
print()
print(sr1.add(sr2))
print()
print(sr1.add(sr2, fill_value = 0))
```
