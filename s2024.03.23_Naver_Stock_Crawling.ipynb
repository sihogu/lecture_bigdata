```python
import pandas as pd
import requests
import os
from bs4 import BeautifulSoup
from time import sleep
from matplotlib import font_manager, rc
import matplotlib.pyplot as plt
import warnings
import folium

# 경고 메시지 생략(무시)
warnings.filterwarnings('ignore')

# 그래프에서 한글처리 문제 해결
font_path = 'malgun.ttf'
font_name = font_manager.FontProperties(fname = font_path).get_name()
rc('font', family = font_name)

# 그래프에서 음수(minus) 기호 문제 해결
plt.rcParams['axes.unicode_minus'] = False
```

```python
def extract_1day(tr):
    td_list = tr.find_all('td')
    del td_list[2]
    date = td_list[0].text.replace('.', '-')
    row = [date]
    for td in td_list[1:]:
        value = td.find('span').text.replace(',', '')
        row.append( int(value) )
    return row
```

```python
code = '005930'
page = 1
url = f'https://finance.naver.com/item/sise_day.naver?code={code}&page={page}'

response = requests.get(url, headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0 x64)'})
print('상태코드:', response.status_code)
print('컨텐츠-타입:', response.headers['content-type'])
```

```python
text = response.text

html = BeautifulSoup(text, 'html.parser')
tr_list = html.find_all('tr', {'onmouseover':'mouseOver(this)'})

rows = []
for tr in tr_list:
    row = extract_1day(tr)
    rows.append(row)

stock = pd.DataFrame(rows, columns = ['date', 'close', 'open', 'max', 'min', 'volume'])
stock = stock.set_index('date')
stock = stock.sort_index(ascending = True)
stock.to_csv('csv/삼성전자_주식.csv', encoding = 'utf-8')
stock
```

```python
code = '005930'

rows = []
for page in range(1, 6):
    url = f'https://finance.naver.com/item/sise_day.naver?code={code}&page={page}'
    response = requests.get(url, headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0 x64)'})
    text = response.text
    html = BeautifulSoup(text, 'html.parser')
    tr_list = html.find_all('tr', {'onmouseover':'mouseOver(this)'})

    for tr in tr_list:
        row = extract_1day(tr)
        rows.append(row)
    sleep(2)
rows
```

```python
stock = pd.DataFrame(rows, columns = ['date', 'close', 'open', 'max', 'min', 'volume'])
stock = stock.set_index('date')
stock = stock.sort_index(ascending = True)
stock.to_csv('csv/삼성전자_주식.csv', encoding = 'utf-8')
stock
```
