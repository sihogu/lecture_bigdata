```python
import pandas as pd
```

**CSV파일**

```python
#행을 열 이름으로 설정
pd.read_csv('sample.csv') #csv파일을 데이터프레임화 시키는 함수
pd.read_csv('sample.csv', header = 1) #인덱스 번호 n 행을 열 이름으로 지정하고, 나머지 위의 행들 제거. 기본값 = 0.
pd.read_csv('sample.csv', header = None) #열 이름을 행으로 하지 않고, 인덱스 번호로 함.

#열을 행 이름으로 설정
pd.read_csv('sample.csv', index_col=False) #기본값. 행 인덱스 명을 지정하지 않음.
pd.read_csv('sample.csv', index_col='c0') # 행 인덱스명을 첫째 행의 이름으로부터 받아와서, 해당 열을 행 인덱스 이름으로 설정.
```

**Excel 파일**

```python
df1 = pd.read_excel('남북한발전전력량.xlsx') #header, index_col 옵션은 동일하다.
df2 = pd.read_excel('남북한발전전력량.xlsx', engine = 'openpyxl') #pandas 기본 엔진이 오류날 경우. 'openpyxl'엔진으로 교체

print(df1.iloc[:,0:6])
print('\n')
print(df2.iloc[:,0:6])
```
