```python
import pandas as pd
import seaborn as sns
import warnings
from numpy import nan
```

```python
titanic = sns.load_dataset('titanic')
titanic.head(4)
```

```python
df = titanic.iloc[:5][['age', 'fare']]
df
```

```python
# 1개의 숫자가 공급되면, 그 수에 10을 더해 반환한다.
def add_10(n):
    return n + 10

# 시리즈.apply(함수) : 시리즈의 개별 원소단위로 함수에 공급하고, 반환받은 값들을
#                      다시 동일한 위치에 넣어서, 원래의 시리즈와 동일한 모양의
#                      시리즈 자료형을 만들어준다.
print(df['age'])
print()
print(df['age'].apply(add_10))
```

```python
def grade(score):
    if score >= 90:   grd = 'A'
    elif score >= 80: grd = 'B'
    elif score >= 70: grd = 'C'
    elif score >= 60: grd = 'D'
    else:             grd = 'F'
    return grd

students = pd.read_csv('csv/class_1.csv', encoding = 'euckr', header = None,
                       names = ['name', 'addr', 'kor', 'eng', 'math', 'sci', 'his'],
                       index_col = 'name')

students['avg'] = students.loc[:, 'kor':].mean(axis = 'columns')

# 시리즈.apply(함수) : 시리즈의 개별원소단위 함수에 공급하고, 반환받은 값들을
#                      다시 동일한 크기(모양)의 시리즈로 반환
students['grade'] = students['avg'].apply(grade)
students
```

```python
def add_three(x, y = 0, z = 0):
    return x + y + z

print(df['age'])
print()
print( df['age'].apply(add_three, y = 10, z = 10) )
print()
print( df['age'].apply(lambda x: x + 10) )
```

```python
# 시리즈.apply(함수)       : 시리즈의 개별 원소단위로 함수에 공급 -> 시리즈
#                            시리즈에 적용할 함수는 원소단위로 공급
# 데이터프레임.apply(함수) : 컬럼(행) 단위로 함수에 공급 -> 시리즈
#                            데이터프레임의 적용할 함수는 시리즈단위로 공급
# 데이터프레임.applymap(함수) : 데이터프레임의 원소단위로 함수에 공급 -> 데이터프레임
#                               데이터프레임의 적용할 함수는 원소단위로 공급
def add_10(data):
    return data + 10

df.applymap(add_10)
```

```python
students.loc[:, 'kor':'his'].applymap(grade)
```

```python
# 시리즈가 인자로 공급되면, 전체 시리즈에서 NaN(누락치) 데이터의 수량을 반환
# 시리즈 -> 시리즈
def missing_values(sr):
    return sr.isnull()

# 시리즈 -> 값
def missing_values_count(sr):
    return sr.isnull().sum()


df = titanic.iloc[15:30][['age', 'fare']]
# 데이터프레임.apply(함수, axis = 'index') : index들을 묶어서 컬럼단위로 함수에 공급
df.apply(missing_values, axis = 'index')
```

```python
df.apply(missing_values_count, axis = 'index')
```

```python
df.apply(missing_values_count, axis = 'columns')
```

```python
students = pd.read_csv('csv/class_1.csv', encoding = 'euckr', header = None,
                       names = ['name', 'addr', 'kor', 'eng', 'math', 'sci', 'his'],
                       index_col = 'name')
students
```

```python
def average(sr):
    return sr.sum() / len(sr.index)

students2 = students.copy()
students2['avg'] = students2.loc[:, 'kor':].apply(average, axis = 'columns')
students2
```

```python
def add_two_obj(sr1, sr2):
    return sr1 + sr2

df2 = df.copy()
df2['ten'] = 10

# 데이터프레임(df2)를 컬럼(axis = 'columns')을 묶어서 행(index)단위로 함수 공급
# Series({}'age':55.0, 'fare':16.0, 'ten':10})
df2['age+ten'] = df2.apply(lambda sr: sr['age'] + sr['ten'], axis = 'columns')
df2
```

```python
# 학생데이터(Series)가 공급되면 국어(kor), 영어(eng), 수학(math) 과목에
# 대해서 합계를 구한 함수
def sub3_average(sr):
    return (sr['kor'] + sr['eng'] + sr['math']) / 3

students2 = students.copy()
print( students2.apply(sub3_average, axis = 'columns') )
students2.apply(lambda sr: (sr['kor'] + sr['eng'] + sr['math']) / 3, axis = 'columns')
```

```python
# 데이터프레임.pipe(함수) : 데이터프레임을 함수에 인자를 공급
# 함수(데이터프레임)
def missing_value(df):
    return df.isnull()

def missing_count(df):
    return df.isnull().sum()

def missing_total_count(df):
    return df.isnull().sum().sum()

df = titanic.loc[15:24, ['age', 'fare', 'deck']]
missing_count(df)
```
