```python
grouped = df.groupby('class')
for key, group in grouped:
    print(key, len(group))

# 그룹핑객체.filter(조건함수) : 조건에 부합되는 데이터프레임들을
#                               결합하여 1개의 데이터프레임
grouped_filter = grouped.filter(lambda df: len(df) >= 200)
grouped_filter
```

```python
# filter(함수) 는 데이터프레임 객체을 전달받아 True / False 으로 반환해야한다.
def over_200(df):
    return len(df) > 200

grouped.filter(over_200)
```

```python
print(grouped['age'].mean())

grouped.filter(lambda df: df['age'].mean() < 30)
```

```python
age_filter = grouped.apply(lambda df: df['age'].mean() < 30)

df_list = []
for x in age_filter.index:
    if age_filter[x] == True:
        age_filter_df = grouped.get_group(x)
        #print('* key:', x)
        #print(age_filter_df.head(3))
        df_list.append(age_filter_df)
        #print()

pd.concat(df_list)
```

```python
titanic2 = titanic.copy()

# who 컬럼을 기준으로 그룹화하고, 그룹별 age컬럼의 평균 나이를 조회(시리즈)
who_mean_age = titanic2.groupby('who')['age'].mean()
print(who_mean_age)
print(who_mean_age['child'])
```

```python
mask = titanic2['who'] == 'child'
df = titanic2[mask]
df['age'] = df['age'].fillna(who_mean_age['child'])

mask = titanic2['who'] == 'man'
df = titanic2[mask]
df['age'] = df['age'].fillna(who_mean_age['man'])

mask = titanic2['who'] == 'woman'
df = titanic2[mask]
df['age'] = df['age'].fillna(who_mean_age['woman'])
```

```python
df_list = []
for who in who_mean_age.index:
    mask = titanic2['who'] == who
    df = titanic2[mask]
    df['age'] = df['age'].fillna(who_mean_age[who])
    df_list.append(df)
dfs = pd.concat(df_list)
dfs.sort_index(inplace = True)
ori_age = titanic2['age']
dfs.insert(4, 'ori_age', ori_age)
dfs[25:50]
```
