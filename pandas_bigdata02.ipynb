```python
import pandas as pd
import seaborn as sns #데이터 내장 라이브러리
```

**시리즈연산**

```python
student1 = pd.Series({'국어':100,'수학':90.5,'영어':80})

student1 / 200 #시리즈의 상수 값 전부 연산(+,-,*,/). 단, 값들이 전부 숫자여야한다.

student2 = pd.Series({'도덕':100,'국어':90.5,'영어':80})

student1 + student2 #시리즈와 시리즈를 연산(+,-,*,/). 단, 값들이 전부 숫자이고 key가 서로 똑같아야한다. 이때 위치가 달라도 같은 key끼리 연산.

#연산자 메소드(add,sub,mul,div)
student1.add(student2, fill_value=0) #연산자 메소드에서 (fill_value=0)으로 비어있거나 맞지않은 key여도 계산을 0과 하여 NaN이 발생되지않게함.
```

**데이터프레임 연산**

```python
titanic = sns.load_dataset('titanic') #'titanic' 데이터셋 불러오기
df = titanic.loc[:,['age','fare']] #'age', 'fare' 열만 인덱싱
df = df.head() #위에서부터 설정된만큼 행 호출, 기본값은 5

df = df + 10 #데이터프레임 내의 상수 값 전부 연산. 값들이 전부 숫자여야한다.

df2 = titanic.loc[:,['age','fare']]
df2 = df2.head()

print(df)
print(df2)
print(df+df2) #데이터프레임과 데이터프레임을 연산. 값들이 전부 숫자이고 key가 서로 똑같아야한다. 없거나 다르면 NaN 발생


#연산자 메소드(add,sub,mul,div)
df.add(df2)
```
