```python
stock = pd.read_csv('csv/삼성전자_주식.csv')
stock['date'] = pd.to_datetime(stock['date'])
stock.info()
```

```python
stock['date'] = stock['date'].astype(str)
dates = stock['date'].str.split('-')
stock['year'] = dates.str.get(0)
stock['year'] = stock['year'].astype(int)
stock['month'] = dates.str.get(1)
stock['month'] = stock['month'].astype(int)
stock['day'] = dates.str.get(2)
stock['day'] = stock['day'].astype(int)
stock.dtypes
```

```python
def str_to_year(text):
    return int(text.split('-')[0])
def str_to_month(text):
    return int(text.split('-')[1])
def str_to_day(text):
    return int(text.split('-')[2])

def str_to_date(text):
    date = text.split('-')
    year = int(date[0])
    month = int(date[1])
    day = int(date[2])
    return pd.Series({'year':year, 'month':month, 'day':day})

stock = pd.read_csv('csv/삼성전자_주식.csv')
#stock['연'] = stock['date'].apply(str_to_year)
#stock['월'] = stock['date'].apply(str_to_month)
#stock['일'] = stock['date'].apply(str_to_day)
stock = pd.concat([stock, stock['date'].apply(str_to_date)], axis = 'columns')
stock
```

```python
titanic = sns.load_dataset('titanic')
titanic.head(5)
```

```python
df = titanic.iloc[0:5]
df
```

```python
mask1 = df['sex'] == 'female'
mask2 = df['fare'] < 10
df[mask1 & mask2]
```

```python
mask = (df['sex'] == 'female') & (df['fare'] < 10)
df[mask]
```

```python
mask = (titanic['age'] >= 10) & (titanic['age'] < 20)
titanic[mask]
```

```python
# 탑승도시(embark_town)가 'Queenstown'인 성년/미성년(who)이 'child'인 승객만 조회
mask = (titanic['embark_town'] == 'Queenstown') & (titanic['who'] == 'child')
titanic[mask]
```

```python
# 10세미만 승객이면서 여성만 조회
mask = (titanic['age'] < 10) & (titanic['sex'] == 'female')
titanic[mask]
```

```python
# 동승객(sibsp)이 3명 ~ 5명(3, 4, 5)인 승객들만 조회
mask1 = (titanic['sibsp'] == 3) | (titanic['sibsp'] == 4) | (titanic['sibsp'] == 5)
mask2 = titanic['sibsp'].isin([3, 4, 5])
titanic[mask2]
```

```python
# 동승객(sibsp)이 3명 ~ 5명(3, 4, 5)인 승객들 중 남성(male) 승객만 조회
mask = titanic['sibsp'].isin([3, 4, 5]) & (titanic['sex'] == 'male')
titanic[mask]
```
