```python
import pandas as pd
```

```python
#########read_csv 함수의 옵션#########
#encoding = 'utf-8', 'ansi', 'euckr' : 문자가 저장된 인코딩 방식을 지정한다. 기본값 = 'utf-8' 
#sep = '구분자'                      : 값들을 구분하는 문자를 지정한다. 기본값 = ',' \t나 \n 같은 중복 연속되는 문자열을 변경해준다.
#header = 'index행' 혹은 'index명'   : 열 이름을 행 index번호(이름)의 행으로 설정한다. 기본값 = 0
#header = None                       : 원래 파일에 columns(필드)명을 포함하지 않는 경우, 첫 행의 열 이름화가 안이루어진다.
#index_col = 'col명'                 : 데이터프레임 생성시 특정 column을 index(축)으로 설정한다.
#names = [col 배열]                  : 데이터프레임 객체 생성시 별도의 columns명들을 지정할 수 있다.
#                                      set_index('col명') 생략 가능
#skiprows = 행의 개수                : csv파일을 읽을 때 행의 n개만큼 생략한다. 보통 csv파일 앞쪽 배열아닌 설명을 생략한다.

#-csv (Comma Seperated Values)
#-tsv (Tab Seperated Values)

#header, index_col, skiprows


# 기존 csv파일에는 columns명으로 사용할 행이 없으므로 header = None
# 그리고 사용자가 임의로 columns명들을 지정하고 names = [columns 배열]
# 그 중 columns(필드)들 중 '이름' 컬럼을 index(축)으로 지정했다.

filename = 'csv/exam5.tsv'
df = pd.read_csv(filename, sep = '\t', header = 0)
df
```

```python
#함수 정의하기
def average(scores):
    avg = sum(scores) / len(scores)
    return avg

def grade(avg):
    if avg >= 90: return 'A'
    elif avg >= 80: return 'B'
    elif avg >= 70: return 'C'
    elif avg >= 60: return 'D'
    elif avg >= 50: return 'E'
    else: return 'F'

    
    
scores = [90,86,97,20,70]
print(grade(average(scores)))
```

```python
#name 컬럼을 index 축으로 지정하고, 국어~역사 점수까지 조회하기
#df1 = df.set_index('name')
#df1.drop('addr',axis = 1)
#df2.apply(grade, axis = 'columns')
#df.apply(정의한 함수명, axis = '축이름')      : 특정 열과 행에 정의한 함수를 전부 전달한다.

df1 = df.copy()
df2 = df1.loc[:,'kor':].apply(average, axis = 'columns')
df['avg'] = df2
df['grd'] = df['avg'].apply(grade)
df.to_csv('csv/result.csv',encoding='ansi') #데이터 프레임을 csv 파일로 원하는 경로에 저장하는 함수
```
