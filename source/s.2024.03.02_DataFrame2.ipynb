```python
import pandas as pd
```

```python
exam_data = {'수학':[90,80,70,60,70],
            '영어':[98,89,99,90,100],
             '음악':[85,95,100,90,90],
            '체육':[100,90,90,80,70]}
df = pd.DataFrame(exam_data, index = ['서준','우현','인아','진수','진희'])
df
```

```python
#열 → 행
df['수학']['서준'] #df['열이름']['행이름']
df.수학['서준'] # df.열이름['행이름']

#행 → 열
df.loc['진수']['영어']
df.loc['진수','영어']
df.iloc[0][0]
df.iloc[0,0]
```

```python
#다수 인덱싱, 원하는 것만 한줄씩
df.loc[['서준','인아'],['음악','체육']]
#연속 인덱싱, 처음:끝-1까지
df.iloc[0:-1,1:-1]
```

```python
df2 = df.copy()
df2['음악'] = [1,2,3,4,5] #해당 열의 원소 전부 변경(혹은 추가). 대신 해당 열의 원본 원소 개수와 같아야한다.
df2['도덕'] = 0
df2
```

```python
#names 옵션으로 열 이름들을 정할 수 있다.
df4 = pd.read_csv('csv/class_1.csv', encoding='euckr', names = ['name','addr','kor','eng','math','sci','his'])
df4.set_index(['name','addr']) #해당하는 열들을 다수 인덱스로 변경
df4.set_index(['name'],inplace=True) #해당하는 열을 단일 인덱스로 변경
df4
```
